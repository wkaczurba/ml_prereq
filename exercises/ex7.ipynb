```python
import numpy as np
import matplotlib.pyplot as plt

N = 100

distance = 20
def dataset1(distance):
    y = np.random.rand(N) * distance*2 - distance
    x = np.sqrt(abs(y ** 2 - distance ** 2))
    x = x * np.random.choice([-1, 1], x.shape)
    return [x,y] 

x, y = dataset1(20)

plt.scatter(x, y)
```
