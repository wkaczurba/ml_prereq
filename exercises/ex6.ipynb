```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

N = 1000

arr = np.random.random([N, 2]) - 0.5 
res = (arr[:,0] * arr[:,1]) > 0

red  = arr.T[:, res].T
blue = arr.T[:, res == False].T
plt.scatter(red[:,0], red[:,1])
plt.scatter(blue[:,0], blue[:,1])
```

```python
# another approach of the same:
arr = np.random.random([N, 2]) - 0.5 
#res = (arr[:,0] * arr[:,1]) > 0
res = np.where( (arr[:,0] * arr[:,1]) > 0, 'b', 'r')
plt.scatter(arr[:,0],arr[:,1], c=res)
```
