```python
# COntinue to work with the MNIST dataset
# COmmon C Programming exercise (with for loops and array indexing)
# Write a function that flips an image 90 degrees clockwise
# Try both the "for loop method" and by making use of Numpy functions

# TODO: Get the MNIST dataset (???)

import mnist
import numpy as np
from PIL import Image
import maptplotlib.pyplot as plt
mnist.init()
```

```python
x_train, t_train, x_test, t_test = mnist.load()

arr = x_train[10,:].reshape(28,28)

def rotate_minus_90(arr):
    rotated = np.zeros(arr.shape)
    max_x, max_y = arr.shape
    for y in range(0, max_y):
        for x in range(0, max_x):
            rotated[y,x] = arr[max_y-x-1,y] #-- [0, 0] -> [maxy, 0], [0, 1] -> [maxy-1, 0]
    return rotated

def rotate_90(arr):
    rotated = np.zeros(arr.shape)
    max_x, max_y = arr.shape
    for y in range(0, max_y):
        for x in range(0, max_x):
            rotated[max_y - y - 1,max_x - x - 1] = arr[max_y-x-1,y] #-- [0, 0] -> [maxy, 0], [0, 1] -> [maxy-1, 0]
    return rotated

rotated = rotate_90(arr)
img = Image.fromarray(rotated) # First image in the training set.
img = img.resize((200,200))

fig, ax = plt.subplots(figsize=(15, 5))
plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
#img.show() # Show the image

img = Image.fromarray(arr) # First image in the training set.
img = img.resize((200,200))
#img.show() # Show the image
plt.subplot(1,3,2)
plt.imshow(img, cmap='gray')

img = Image.fromarray(np.rot90(arr)) # First image in the training set.
img = img.resize((200,200))
plt.subplot(1,3,3)
plt.imshow(img, cmap='gray')
```
